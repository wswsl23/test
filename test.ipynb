### 
1. 从链接页面获取公开数据
2. 需要获取数据的条件: Bond Type=Treasury Bond, Issue Year=2023
3. 解析返回表格数据，列名包括ISIN, Bond Code, Issuer, Bond Type, Issue Date, Latest Rating
3. 保存成有效csv文件，


```python
from bs4 import BeautifulSoup
import requests
import json
import xlsxwriter as xw
import pandas as pd

headers = { # 请求头
'Referer':'https://iftp.chinamoney.com.cn/english/bdInfo/',
'Cookie':'apache=bbfde8c184f3e1c6074ffab28a313c87; _ulta_id.ECM-Prod.ccc4=cd9586d12cf09699; _ulta_ses.ECM-Prod.ccc4=580f37bc5a4f942a; AlteonP10=ABD5aiw/F6yWk0AF7ZSvKQ$$',
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36',
}
url1="https://iftp.chinamoney.com.cn/ags/ms/cm-u-bond-md/BondMarketInfoListEN"
cc=[]
for i in range(1,5):
    payload={ # 表单
    'pageNo': f'{i}', # 页数 共4页
    'pageSize': '15',
    'bondType': '100001',
    'issueYear': '2023',
    }
    response_1=requests.post(url1,data=payload,headers=headers)
    soup = BeautifulSoup(response_1.text,'html.parser')
    data = eval(soup.text.replace('null','None'))
    for x in data['data']['resultList']:
        cc.append([x['isin'],x['bondCode'],x['entyFullName'],x['bondType'],x['issueStartDate'],x['debtRtng']])

df = pd.DataFrame(cc,columns=['ISIN', 'Bond Code', 'Issuer', 'Bond Type', 'Issue Date', 'Latest Rating'])
df.to_csv('爬虫.csv',index=False)
```
